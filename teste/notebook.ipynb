# Teste dos Endpoints

Notebook para testar os endpoints de upload e métricas.

```python
import requests
import json

BASE_URL = "http://localhost:8000"
```

```python
## 1. Upload
```

```python
# Ajuste o caminho para o seu arquivo CSV/XLSX
FILE_PATH = "dados.xlsx"

with open(FILE_PATH, "rb") as f:
    resp = requests.post(
        f"{BASE_URL}/files/upload",
        files={"file": (FILE_PATH.split('/')[-1], f)},
    )

print(resp.status_code)
upload_result = resp.json()
print(json.dumps(upload_result, indent=2))

# Salvar o ingestion_id para usar nos próximos testes
ingestion_id = upload_result["ingestion_id"]
print(f"\ningestion_id: {ingestion_id}")
```

## 2. Métricas Consolidadas

```python
resp = requests.post(
    f"{BASE_URL}/metrics/",
    json={"ingestion_id": ingestion_id},
)

print(resp.status_code)
print(json.dumps(resp.json(), indent=2))
```

## 3. Métricas Disponíveis

```python
resp = requests.get(f"{BASE_URL}/metrics/available")

print(resp.status_code)
print(json.dumps(resp.json(), indent=2))
```

## 4. Evolução Temporal

```python
# Todas as métricas de receita juntas
resp = requests.post(
    f"{BASE_URL}/metrics/temporal",
    json={
        "ingestion_id": ingestion_id,
        "metric_names": ["receita_total", "receita_real", "receita_inadimplente"],
        "freq": "M",
        "mode": "pontual",
    },
)

print(resp.status_code)
print(json.dumps(resp.json(), indent=2, default=str))
```

```python
# Ticket médio
resp = requests.post(
    f"{BASE_URL}/metrics/temporal",
    json={
        "ingestion_id": ingestion_id,
        "metric_names": ["ticket_medio"],
        "freq": "M",
        "mode": "acumulativo",
    },
)

print(resp.status_code)
print(json.dumps(resp.json(), indent=2, default=str))
```

```python
# Taxa de inadimplência
resp = requests.post(
    f"{BASE_URL}/metrics/temporal",
    json={
        "ingestion_id": ingestion_id,
        "metric_names": ["taxa_inadimplencia"],
        "freq": "M",
        "mode": "a",
    },
)

print(resp.status_code)
print(json.dumps(resp.json(), indent=2, default=str))
```

## 5. Filtros

```python
resp = requests.post(
    f"{BASE_URL}/filters",
    json={
        "ingestion_id": ingestion_id,
    },
)

print(resp.status_code)
print(json.dumps(resp.json(), indent=2, default=str))
```

## 6. Métricas (Com filtros)

```python
resp = requests.post(
    f"{BASE_URL}/metrics/",
    json={
  "ingestion_id": ingestion_id,
  "filter_criteria": {
    "valor": { "min": 100, "max": 5000 },
    "data": {
      "min": "2024-01-01T00:00:00",
      "max": "2025-12-01T00:00:00"
    },
    #"status": ["pago", "pendente"],
    "cliente": ["empresa a", "startup x"],
    #"is_pago": [True],
    "empresa": [True],
    "recorrencia": ["recorrente"]
  }
},
)

print(resp.status_code)
print(json.dumps(resp.json(), indent=2))
```
